```python
import pandas as pd
df_temp = pd.read_csv("PairedCoursesAndClasses.csv")
df = pd.DataFrame()
df['Courses'] = df_temp['Courses']
df['Classes'] = df_temp['Classes']
df
```

```python
department = []
course_num = []

for course in df['Courses']:
    dep = ""
    x = 0
    while(not course[x].isdigit()):
        dep = dep + course[x]
        x += 1
    department.append(dep[1: len(dep)-1])
    course_num.append(course[x:x+4])
print(department)
print(course_num)
```

```python
class_num = []
section_num = []
class_type = []
credits = []
seat_count = []
end_string = []
for class_str in df['Classes']:
    class_num.append(class_str[:5])
    x = 6
    sec_num = ""
    while(class_str[x].isdigit()):
        sec_num += class_str[x]
        x+=1
    section_num.append(sec_num)
    c_type = ""
    while(not class_str[x] == "("):
        c_type += class_str[x]
        x+=1
    class_type.append(c_type[:len(c_type)-1])
    credit = ""
    while(not class_str[x] == ")"):
        credit += class_str[x]
        x+=1
    credits.append(credit[1:])
    x+=1
    if(class_str[x] == 'W' or class_str[x:x+8] == "Open, WL"):
        while(not class_str[x] == ")"):
            x+=1
    while(not class_str[x] == "/"):
        x+=1
    x+=2
    seats = ""
    while(class_str[x].isdigit()):
        seats += class_str[x]
        x+=1
    seat_count.append(seats)
    end_string.append(class_str[x:])
print(end_string)
```

```python
meeting_place = []
remainder = []
z = 1
for class_str in end_string:
    z += 1
    x = len(class_str) - 1
    if(z == 6252):
        meeting_place.append("N Grounds Rec Ctr Pool - Deep")
    elif(class_str[x-20:x] == "Desktop-Asynchronous"):
        meeting_place.append("Desktop-Asynchronous")
        x-=21
    elif(class_str[x-18:x] == "Contact Department"):
        meeting_place.append("Contact Department")
        x-=19
    elif(class_str[x-3:x] == "TBA"):
        meeting_place.append("TBA")
        x -= 4
    else:
        meeting_pl = ""
        while (not (class_str[x-1:x+1] == 'pm' or class_str[x-1:x+1] == 'am')):
            meeting_pl = class_str[x] + meeting_pl
            x -= 1
        meeting_place.append(meeting_pl[:len(meeting_pl)-1])
    remainder.append(class_str[:x+1])
print(remainder)
```

```python
weekdays = ["Mo", "Tu", "We", "Th", "Fr", "Sa", "Su"]
meeting_time = []
remaining_chunk = []
for class_str in remainder:
    x = len(class_str) - 1
    if(class_str[x-2:] == "TBA"):
        meeting_time.append("TBA")
        x -= 3
    else:
        meet_t = ""
        while(not ((class_str[x-1:x+1] in weekdays) or (class_str[x-2:x+1] == "TBA"))):   
            #print(class_str[x-2:x+1])
            meet_t = class_str[x] + meet_t
            x -= 1
        meeting_time.append(meet_t)
    remaining_chunk.append(class_str[:x+1])
print(meeting_time)
```

```python
meeting_dates = []
professor = []
for class_str in remaining_chunk:
    x = len(class_str) - 1
    if(class_str[x-2:] == "TBA"):
        meeting_dates.append("TBA")
        x-=3
    else:
        meet_d = ""
        while(class_str[x-1:x+1] in weekdays):
            meet_d = class_str[x-1:x+1] + meet_d
            x-=2
        meeting_dates.append(meet_d)
    professor.append(class_str[:x+1])
print(len(remaining_chunk))
```

```python
df_final = pd.DataFrame()
df_final["Department"] = department
df_final["Course_Num"] = course_num
df_final["Class_Num"] = class_num
df_final["Section_Num"] = section_num
df_final["Class_Type"] = class_type
df_final["Credits"] = credits
df_final["Seat_Count"] = seat_count
df_final["Meeting_Place"] = meeting_place
df_final["Meeting_Time"] = meeting_time
df_final["Meeting_Dates"] = meeting_dates
df_final["Professor"] = professor
```
