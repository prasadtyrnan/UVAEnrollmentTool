```python
import pandas as pd
def isClass(string):
    for x in range(0,5):
        if(not string[x].isdigit()):
            return False
    return True
def isCourse(string):
    if(len(string)<2):
        return False
    x = 0
    if(";" in string):
        return False
    while(x<len(string) and not string[x].isdigit()):
        x = x+1
    if(x>2 and x<len(string)):
        for i in range(x,x+4):
            if(not string[i].isdigit()):
                return False
        if(string[1:3].isupper()):
            return True
    return False
def isDepartment(string):
    if (not (any(char.isdigit() for char in string))):
        if (string[:1] != " "):
            return True
    return False
```

```python
courses = []
classes = []
f = open("RawCourseList.txt", "r")
last_course = "";
for string in f:
    if(isCourse(string)):
        last_course = string;
    elif(isClass(string)):
        courses.append(last_course)
        classes.append(string)
```

```python
df = pd.DataFrame()
df['Courses'] = courses
df['Classes'] = classes
df
```

```python
df.to_csv(r"PairedCoursesAndClasses.csv")
```
